# Histogram of Oriented Gradients (HOG)
An example from https://scikit-image.org/docs/dev/auto_examples/features_detection/plot_hog.html

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure
```

```python
def normalize(data):
    return ((data - data.min()) / (data.max() - data.min())).astype(np.float32)
```

```python
image = data.astronaut()
image = normalize(image)
print(image.shape, image.dtype)
```

```python
plt.figure()
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
fd, hog_image = hog(image, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
```

```python
# Min-Max Normalization
hog_image = (hog_image - hog_image.min()) / (hog_image.max() - hog_image.min())
hog_image = hog_image.astype(np.float32)
```

```python
plt.figure()
plt.imshow(hog_image, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
print(hog_image.min(), hog_image.max())
print(hog_image.shape, hog_image.dtype)
```
